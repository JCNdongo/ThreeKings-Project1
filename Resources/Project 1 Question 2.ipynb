```python
import os 
import csv
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import smtplib
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
```

```python
#Load the dataset
file = os.path.join('Desktop', 'GitHub', 'ThreeKings-Project1', 'Resources', 'wage_predict.csv')
data = pd.read_csv(file) 
```

```python
data.head()
```

```python
# Regression analysis of influencing on salary
X = sm.add_constant(data['influencing'])
model = sm.OLS(data['monthly_salary'], X)
results = model.fit() 
print(results.summary())
```

```python
# Scatter plot with regression line
plt.scatter(data['influencing'], data['monthly_salary'])
plt.plot(data['influencing'], results.predict(), color='red')
plt.xlabel('influencing')
plt.ylabel('Salary')
plt.title('Influencing vs. Salary')
plt.show()
```
